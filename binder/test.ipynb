```python
def grapha(x, y):
    fig, ax = plt.subplots(figsize=(10,5))
    ax.bar(df[x], df[y])
    plt.show()
widgets.interact(grapha, x=['G3', 'Mjob','Fjob','age'],y=['Dalc', 'Walc']);
```

The graph above shows the relationship of

absences - number of school absences
Mjob - mother's job
Fjob - father's job
G3 - final grade (numeric: from 0 to 20, output target)

with 
workday alcohol consumption (Dalc) and weekday alcohol consumption(Walc). 

It shows that studnts with highest scores had 0 to 1 drinks during workdays.. It also shows that students who drank the most got score from 5 - 15. The number of drinks significantly decreases for students who got over 15. However, this is not completely true as students who drank as little as 2 also got scores from 5 - 10.

One thing I found interesting was that almost all students drank at level 5 for weekday consumptions, except the ones who got about 18 - 20 score.  


Another interesting thing I found about this dataset was that students whose parents' job is in health field tend to drink less during workdays. I think this could be because their parents might be more cautious about what their child is consuming. 


For the age, the interesting thing was that students of age 21 decreased in drinking significantly while the graph below shows American college students' grade drop at the age 21. Scholars says that this is because the legal age in United States is 21, and the grade starts to drop when they start drinking. For Portugies students, their drinking age is 18. At 21, we can see that their alcohol consumption decreased.
